```python
fft_image = np.fft.fft2(img)

fft_shift = np.fft.fftshift(fft_image)
```

```python
magnitude_spectrum = np.log(np.abs(fft_shift) + 1)

plt.figure(figsize=(8,8))
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title("Espectro de Frecuencias")
plt.axis('off')
plt.show()
```

```python
#aqui se pueden eliminar las frecuencias altas ya que en la imagen anterior vemos cual es el dominio de fourier
rows, cols = img.shape

crow, ccol = rows // 2, cols // 2

mask = np.zeros((rows, cols))

radio = 50

for i in range(rows):
    for j in range(cols):
        if (i-crow)**2 + (j-ccol)**2 <= radio**2:
            mask[i,j] = 1
```

```python
plt.imshow(mask, cmap='gray')
plt.title("Filtro Pasa Bajas")
plt.axis('off')
plt.show()
```

```python
#aplicando el filtro
filtered_fft = fft_shift * mask
```

```python
#recontruimos la imagen
ifft_shift = np.fft.ifftshift(filtered_fft)

reconstructed = np.fft.ifft2(ifft_shift)

reconstructed = np.abs(reconstructed)
```

```python
#comparamos los resultado
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(reconstructed, cmap='gray')
plt.title("Filtrada")
plt.axis('off')

plt.show()
```
